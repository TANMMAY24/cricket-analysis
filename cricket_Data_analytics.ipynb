```python
import pandas as pd
import json
```

```python
with open ('/content/t20_wc_match_results.json') as f:
  data=json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
df_match.describe()
```

```python
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
match_ids_dict={}
for index,row in df_match.iterrows():
  key1=row['team1']+' Vs '+ row['team2']
  key2=row['team2']+' Vs '+row['team1']

  match_ids_dict[key1]=row["match_id"]
  match_ids_dict[key2]=row["match_id"]

match_ids_dict

```

```python
with open('/content/t20_wc_batting_summary.json') as f:
  data=json.load(f)
  all_rec=[]
  for rec in data:
    all_rec.extend(rec['battingSummary'])
df_batting=pd.DataFrame(all_rec)
df_batting
```

```python
df_batting["out/not_out"]=df_batting.dismissal.apply(lambda x:"out" if len(x)>0 else "not out")
df_batting.head(11)
```

```python
df_batting
```

```python
df_batting.drop(columns=["dismissal"],inplace=True)
df_batting.head()
```

```python
import re
def remove_special_characters(text):
    pattern = r'[^a-zA-Z0-9\s]'  # Pattern to match any non-alphanumeric characters
    return re.sub(pattern, '', text)
df_batting['batsmanName'] = df_batting['batsmanName'].apply(remove_special_characters)
df_batting
```

```python
df_batting["mach_id"]=df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.to_csv('/content/sample_data/hh.csv',index=False)
```
